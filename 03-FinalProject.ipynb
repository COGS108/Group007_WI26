## Abstract

This project examines how daily COVID-19 case counts vary across days of the week at the county level and how those variations are associated with mobility patterns. Using a large integrated dataset of county-level COVID-19 case counts and Google mobility measures, we investigate whether reported cases differ systematically between weekdays and weekends and whether mobility indicators such as workplace and residential activity help explain changes in case counts. The broader goal is to better understand how reporting patterns and population movement shaped observed pandemic trends.

Our analysis uses a dataset containing 935,443 observations and 17 variables, including daily case counts, date, county identifiers, day-of-week indicators, and several mobility measures. After cleaning and preparing the data, we conducted exploratory analysis and then used a Mann-Whitney U test to compare weekday and weekend case distributions because the case data were highly skewed. We also reported Cohen’s d on log-transformed case counts to assess the size of the weekday-weekend difference. To examine mobility relationships more directly, we fit a series of linear regression models with log-transformed daily cases as the outcome, beginning with a weekend-only baseline model and then adding workplace, residential, retail and recreation, transit, holiday, and time-trend variables.

Our results show a clear and statistically significant weekday-weekend pattern in reported COVID-19 cases, with weekend case counts consistently lower than weekday counts. The Mann-Whitney U test supports this difference at p < 0.001, and the effect size is moderate at approximately Cohen’s d = 0.51. In the regression analysis, the weekend indicator remains negative across all model specifications, while workplace mobility shows a positive association with case counts, which is consistent with the idea that greater movement and interaction are linked to higher transmission. However, residential mobility also shows a positive association in the full model, which does not match our initial expectation and suggests a more complicated relationship between mobility and case trends.

Overall, our findings show that both reporting effects and mobility behavior matter, but the relationships are more nuanced than a simple causal story. The positive residential coefficient and the temporal trends suggest that mobility changes may often reflect behavioral responses to rising cases rather than only causing them. Although adding mobility and time variables improves model performance, the full model still explains less than 20% of the variance in log daily cases, showing that many important influences remain outside the model. These results highlight the importance of accounting for reporting structure, time trends, and broader pandemic dynamics when interpreting COVID-19 case data.

## Background and Prior Work

Mobility rates and COVID-19 case counts exhibited a substantial correlation throughout the pandemic. In response to the rapid spread of the virus, public health authorities urged individuals to remain at home whenever possible, recognizing that increased movement and social interaction heightened the risk of transmission. From the onset of this global crisis, policymakers closely monitored the relationship between population mobility and infection rates in an effort to mitigate uncontrolled surges in cases. A study conducted by the Centers for Disease Control and Prevention (CDC) examined anonymized cell phone GPS data to assess patterns of movement across counties in the United States. The researchers observed that reductions in trips outside the home were associated with subsequent declines in reported COVID-19 cases. Specifically, changes in mobility were analyzed in relation to new case counts approximately 11 days later, reflecting the typical incubation period and reporting delays. The study also compared urban and rural areas to determine whether mobility patterns affected transmission differently across settings. This distinction was critical, as urban centers tend to be more densely populated, exhibit distinct transportation behaviors, and may experience differing timelines of viral introduction and spread compared to rural communities.

For our project, we aim to more closely examine variation in COVID-19 case counts across days of the week, with particular attention to identifying when case spikes are most pronounced. Consistent with our hypothesis, we anticipate that reported cases will be higher during weekdays compared to weekends. One possible explanation is that essential workers continued in-person employment throughout much of the pandemic, increasing opportunities for exposure during the workweek. Workplace environments may facilitate transmission due to repeated close interactions with coworkers and contact with shared surfaces (e.g., counters, equipment, communal appliances). These routine occupational exposures could contribute to elevated transmission risk during weekdays, which may subsequently be reflected in higher reported case counts. This hypothesis is supported by prior empirical work conducted by Hannah Hoffman and The COVID Tracking Project, which examined temporal reporting patterns in state-level COVID-19 case data. Their analysis demonstrated a consistent decline in reported case counts over weekends. Importantly, this pattern was attributed not only to behavioral changes but also to structural factors, including reduced testing, reporting delays, and public health regulations that limited activity during weekends in many states at the time.

In a study conducted by Bulut Boru and M. Emre Gursoy, the authors developed a computational model designed to identify and learn patterns between human mobility and subsequent COVID-19 case counts. Their approach dynamically selects the optimal time frame and predictive method to enhance accuracy. When evaluated across 13 countries, the model demonstrated a high degree of precision in forecasting daily COVID-19 cases. The study further revealed a strong correlation between the movement patterns of individuals and the spread of the virus within their communities. Findings of this nature underscore the substantial impact that human mobility can have on disease transmission, providing empirical justification for the strict travel restrictions and mobility-limiting policies implemented by governments worldwide during the pandemic.

### Analysis 1: Weekday vs. Weekend Differences in Reported COVID-19 Cases

To formally test whether reported COVID-19 case counts differ between weekdays and weekends, we use a Mann-Whitney U test. We choose this non-parametric test instead of a t-test because `daily_cases_clean` is highly skewed and contains many low-value observations, so a normality-based test would be less appropriate. The null hypothesis is that weekday and weekend case counts come from the same distribution.

Because statistical significance is easy to obtain with such a large dataset, we also report an effect size using Cohen’s d on log-transformed case counts. This helps distinguish between a difference that is statistically detectable and one that is substantively large.
Given the very large sample size (over 935,000 observations), even small differences are likely to be statistically significant.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Load directly from Google Drive (bypasses missing local file)
url = "https://drive.google.com/uc?export=download&id=1hQwLJdeMyXGf2Fp0xEBXQrcj8sZsRYca"
df = pd.read_csv(url)
df['date'] = pd.to_datetime(df['date'])

# Clean: clip negatives, create log outcome
df['daily_cases_clean'] = df['daily_cases'].clip(lower=0)
df['log_daily_cases'] = np.log1p(df['daily_cases_clean'])

print(f'Loaded: {df.shape[0]:,} rows, {df.shape[1]} columns')
print(f'Date range: {df["date"].min().date()} to {df["date"].max().date()}')

# Separate weekday and weekend samples
weekday_cases = df.loc[df['is_weekend'] == 0, 'daily_cases_clean'].dropna()
weekend_cases = df.loc[df['is_weekend'] == 1, 'daily_cases_clean'].dropna()

# Mann-Whitney U test (one-sided: weekday > weekend)
u_stat, p_value = stats.mannwhitneyu(weekday_cases, weekend_cases, alternative='greater')

# Cohen's d on log-transformed values
log_weekday = np.log1p(weekday_cases)
log_weekend = np.log1p(weekend_cases)
pooled_std = np.sqrt(
    (log_weekday.std()**2 * (len(log_weekday)-1) + log_weekend.std()**2 * (len(log_weekend)-1))
    / (len(log_weekday) + len(log_weekend) - 2)
)
cohens_d = (log_weekday.mean() - log_weekend.mean()) / pooled_std

print('=== Weekday vs. Weekend: Mann-Whitney U Test ===')
print(f'Weekday N:          {len(weekday_cases):>10,}')
print(f'Weekend N:          {len(weekend_cases):>10,}')
print(f'Weekday mean cases: {weekday_cases.mean():>10.2f}')
print(f'Weekend mean cases: {weekend_cases.mean():>10.2f}')
print(f'Weekday median:     {weekday_cases.median():>10.1f}')
print(f'Weekend median:     {weekend_cases.median():>10.1f}')
print(f'U statistic:        {u_stat:>10,.0f}')
print(f'p-value:            {p_value:.2e}')
print(f"Cohen's d (log):    {cohens_d:>10.4f}")
```

```python
# Figure: Weekday vs Weekend distributions
fig, axes = plt.subplots(1, 2, figsize=(11, 5))

plot_data = df[['is_weekend', 'log_daily_cases']].dropna().copy()
plot_data['Day Type'] = plot_data['is_weekend'].map({0.0: 'Weekday', 1.0: 'Weekend'})

# Violin plot
sns.violinplot(
    data=plot_data, x='Day Type', y='log_daily_cases',
    palette={'Weekday': '#2980b9', 'Weekend': '#c0392b'},
    inner='quartile', ax=axes[0], cut=0
)
axes[0].set_xlabel('Day Type', fontsize=11)
axes[0].set_ylabel('log(1 + Daily Cases)', fontsize=11)
axes[0].set_title('A. Log Daily Cases by Day Type', fontsize=11)

# Boxplot by day of week
dow_plot = df[['day_of_week', 'log_daily_cases']].dropna().copy()
day_labels_map = {0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}
dow_plot['Day'] = dow_plot['day_of_week'].map(day_labels_map)
day_order = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
box_colors = ['#2980b9']*5 + ['#c0392b']*2

bp = axes[1].boxplot(
    [dow_plot.loc[dow_plot['Day']==d, 'log_daily_cases'].values for d in day_order],
    labels=day_order, patch_artist=True, showfliers=False,
    medianprops=dict(color='black', linewidth=1.5)
)
for patch, color in zip(bp['boxes'], box_colors):
    patch.set_facecolor(color)
    patch.set_alpha(0.7)
axes[1].set_xlabel('Day of Week', fontsize=11)
axes[1].set_ylabel('log(1 + Daily Cases)', fontsize=11)
axes[1].set_title('B. Log Daily Cases by Day of Week', fontsize=11)

plt.suptitle(
    'Figure 1. Weekday vs. Weekend Distribution of Log-Transformed Daily COVID-19 Cases\n'
    'Blue = weekdays, red = weekends. Weekday medians are consistently higher across all five weekdays.\n'
    'Mann-Whitney U test: p < 0.001. Outliers suppressed for readability.',
    fontsize=9, y=-0.05, ha='center'
)
plt.tight_layout()
plt.show()
```

Figure 1 caption. Boxplots of log-transformed daily COVID-19 case counts by day of week. Weekdays are shown in blue and weekends in red. Outliers are suppressed for readability.

The Mann-Whitney U test indicates that weekday case counts are significantly higher than weekend case counts (`p < 0.001`). This supports our hypothesis that reported daily cases vary systematically across the week. The effect size (Cohen’s d ≈ 0.51) indicates a moderate difference between weekday and weekend case distributions. However, day-of-week alone still explains only part of the overall variation in county-level case counts. This is consistent with the broader context of the pandemic, where population size, regional waves, testing and reporting practices, and local conditions are also major drivers of case counts.

Importantly, this result should not be interpreted as purely behavioral. The weekend reporting effect likely contributes to the lower weekend values we observe, meaning that the weekday-weekend difference reflects both real social patterns and administrative reporting patterns.

### Analysis 2: Regression Models of Mobility and Daily COVID-19 Cases

To quantify the relationship between mobility behavior and COVID-19 case counts while controlling for day-of-week and time, we fit a series of Ordinary Least Squares (OLS) regression models with `log_daily_cases` as the outcome variable. We use the log-transformed outcome because the raw case counts are highly skewed.

We compare three models of increasing complexity. Model 1 includes only `is_weekend` as a baseline test of weekday-weekend reporting effects. Model 2 adds the two mobility variables most central to our hypothesis: `workplaces` and `residential`. Model 3 adds `retail_recreation`, `transit`, `is_holiday`, and a week-index time trend. We evaluate model performance using `R²`, Mean Absolute Error (MAE), and Root Mean Squared Error (RMSE). These metrics help show both how much variance the model explains and how far predictions tend to be from observed values.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Create week index time trend
df['week_index'] = ((df['date'] - df['date'].min()).dt.days // 7).astype(int)

def fit_ols(feature_cols, label):
    """Fit OLS regression on complete cases for given features.
    
    Parameters
    ----------
    feature_cols : list of str
        Predictor column names.
    label : str
        Model label for display.
        
    Returns
    -------
    dict with model, metrics, and coefficients.
    """
    subset = df[feature_cols + ['log_daily_cases']].dropna()
    X = subset[feature_cols].values
    y = subset['log_daily_cases'].values
    model = LinearRegression()
    model.fit(X, y)
    y_pred = model.predict(X)
    mae_cases = np.mean(np.abs(np.expm1(y) - np.expm1(y_pred)))
    return {
        'label': label,
        'n': len(y),
        'r2': r2_score(y, y_pred),
        'mae_log': mean_absolute_error(y, y_pred),
        'rmse_log': np.sqrt(mean_squared_error(y, y_pred)),
        'mae_cases': mae_cases,
        'coefs': dict(zip(feature_cols, model.coef_)),
        'intercept': model.intercept_,
        'features': feature_cols
    }

m1 = fit_ols(['is_weekend'], 'Model 1: Weekend only')
m2 = fit_ols(['is_weekend', 'workplaces', 'residential'], 'Model 2: Weekend + key mobility')
m3 = fit_ols(
    ['is_weekend', 'workplaces', 'residential', 'retail_recreation', 'transit', 'is_holiday', 'week_index'],
    'Model 3: Full specification'
)

# Print results table
results_table = pd.DataFrame([
    {
        'Model': m['label'],
        'N': f"{m['n']:,}",
        'R²': f"{m['r2']:.4f}",
        'MAE (log)': f"{m['mae_log']:.4f}",
        'RMSE (log)': f"{m['rmse_log']:.4f}",
        'MAE (cases)': f"{m['mae_cases']:.1f}"
    }
    for m in [m1, m2, m3]
])
print(results_table.to_string(index=False))
print()
print('Model 3 Coefficients:')
for feat, coef in m3['coefs'].items():
    print(f'  {feat:25s}: {coef:+.4f}')
```

Figure 2A caption. Comparison of `R²` values across the three regression models.

Figure 2B caption. Coefficient estimates from the full regression specification. Positive coefficients indicate a positive association with log daily cases, while negative coefficients indicate a negative association.

The regression results are mixed. Across all model specifications, the coefficient for `is_weekend` is negative, indicating that weekends are associated with lower reported case counts even after accounting for mobility and time. The coefficient for `workplaces` is positive, which is consistent with our hypothesis that greater workplace mobility is associated with higher COVID-19 case counts. However, the coefficient for `residential` is also positive in the full model, which does not match our original hypothesis. This suggests that once additional predictors and time trends are included, residential mobility may be capturing broader pandemic conditions rather than simply reflecting protective stay-at-home behavior.

The performance metrics also help contextualize these findings. `R²` shows the proportion of variance explained by each model, while MAE and RMSE summarize average prediction error on the log-transformed outcome scale. Although the fuller models improve on the weekend-only baseline, the `R²` values remain modest. This indicates that mobility and day-of-week patterns explain only part of the variation in county-level COVID-19 cases. That result is reasonable, since many important influences on case counts—such as county population size, policy differences, vaccination levels, and variant dynamics—are not fully captured in this dataset.

Workplace mobility shows negative short-lag correlations with average case counts, while residential mobility shows positive short-lag correlations. This likely reflects broader pandemic dynamics: during major surges, workplace mobility decreased and residential mobility increased as people stayed home more. These results reinforce that mobility measures can reflect both behavioral responses to rising cases and potential drivers of transmission, making simple causal interpretation difficult.

## Discusison and Conclusion

Overall, our analysis shows clear but nuanced patterns in how COVID-19 case counts vary across days of the week and how these patterns relate to mobility behavior. We find strong evidence that reported cases are systematically lower on weekends than weekdays. The Mann-Whitney U test confirms that this difference is statistically significant, and the moderate effect size suggests it is not just a trivial artifact of the large dataset. At the same time, mobility variables, particularly workplace and residential mobility, are associated with case counts, though their effects are not always straightforward.

Connecting back to prior work, our findings are consistent with existing literature that highlights a relationship between mobility and COVID-19 transmission. Like the CDC and other studies mentioned in the background, we observe that workplace mobility is positively associated with higher case counts, aligning with the idea that increased movement and interaction facilitate transmission. However, our results extend this work by explicitly incorporating day-of-week variation and showing that reporting structures, not just behavior, play a significant role in observed case patterns.

When we bring together the different analyses, a more complex picture emerges. The regression models show that weekends are consistently associated with lower reported cases even after controlling for mobility and time, reinforcing the presence of a reporting effect. Workplace mobility behaves as expected, with a positive association with cases. However, residential mobility, which we initially expected to be negatively associated with cases, shows a positive coefficient in the full model. Additionally, lagged correlation results show that workplace mobility is negatively correlated with cases at short lags, while residential mobility is positively correlated. These seemingly contradictory findings suggest that mobility is not only a driver of transmission but also a response to it.

This tension in the results likely reflects underlying pandemic dynamics rather than a failure of our hypothesis. During surges, people tend to reduce workplace mobility and increase time at home, meaning that mobility changes often follow rising case counts rather than cause them. This helps explain why residential mobility appears positively associated with cases in the regression: it is capturing behavioral responses to worsening conditions, not the protective effect of staying home itself. In this sense, mobility variables reflect both potential drivers of transmission and reactions to it, making simple causal interpretations difficult.

Another important takeaway is that, although adding mobility variables improves model performance, the overall explanatory power remains modest. Even the most complete model explains less than 20% of the variance in log daily cases. This indicates that many important factors are not captured in our dataset, such as vaccination rates, policy interventions, demographic differences, and the spread of different variants. Day-of-week and mobility clearly matter, but they are only part of a much larger system influencing case counts.

There are several limitations to this study. First, the data are observational and cross-sectional at the county-day level, which limits our ability to make causal claims. Second, COVID-19 case counts are affected by testing availability and reporting delays, especially on weekends, which introduces systematic bias. Third, the mobility data come from smartphone users and may not represent the entire population, particularly in rural or lower-income areas. Missingness in certain mobility variables, such as parks and transit, further reduces reliability. Finally, the strong skewness in case counts required transformation, which may obscure some real-world interpretability.

If this work were to continue, several next steps could strengthen the analysis. Incorporating additional variables, such as vaccination rates, policy measures, and demographic controls, could improve model performance and interpretation. A more explicit time-series approach, such as using lagged regression or causal inference techniques, could better distinguish between mobility as a cause versus a response to case changes. Longitudinal or panel methods that more carefully track changes within counties over time would also help address some of the limitations of the current approach.

In conclusion, we find that COVID-19 case counts exhibit clear weekday-weekend patterns and are meaningfully associated with mobility behavior, but these relationships are complex and not purely causal. Reporting effects, behavioral responses, and broader pandemic dynamics all interact to shape the patterns we observe. While mobility provides useful insight into population behavior during the pandemic, it should be interpreted alongside other factors to fully understand changes in COVID-19 case counts.
